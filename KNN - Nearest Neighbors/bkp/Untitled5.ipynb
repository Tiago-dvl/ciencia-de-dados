```python
import matplotlib.pyplot as plt
import numpy as np

# Dados Originais
idades = np.array([25, 40, 60, 35])
salarios = np.array([3000, 8000, 15000, 4500])

plt.figure(figsize=(10, 6))
plt.scatter(idades, salarios, s=100, color='red', label='Clientes')
plt.title('Dados Originais (Sem Escalonamento)')
plt.xlabel('Idade (anos)')
plt.ylabel('Salário (R$)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xlim(0, 70) # Ajustando limites para visualização
plt.ylim(0, 16000)
for i, txt in enumerate(["001", "002", "003", "004"]):
    plt.annotate(txt, (idades[i], salarios[i]), textcoords="offset points", xytext=(0,10), ha='center')
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

# Dados Originais
idades = np.array([25, 40, 60, 35]).reshape(-1, 1)
salarios = np.array([3000, 8000, 15000, 4500]).reshape(-1, 1)

# Aplicar Padronização
scaler_idade = StandardScaler()
idades_scaled = scaler_idade.fit_transform(idades)

scaler_salario = StandardScaler()
salarios_scaled = scaler_salario.fit_transform(salarios)

plt.figure(figsize=(10, 6))
plt.scatter(idades_scaled, salarios_scaled, s=100, color='blue', label='Clientes Padronizados')
plt.title('Dados Após Padronização (StandardScaler)')
plt.xlabel('Idade Padronizada')
plt.ylabel('Salário Padronizado')
plt.grid(True, linestyle='--', alpha=0.7)
# Ajustando limites para visualização, geralmente em torno de -2 a 2 ou -3 a 3 para dados padronizados
plt.xlim(-2, 2)
plt.ylim(-2, 2)
for i, txt in enumerate(["001", "002", "003", "004"]):
    plt.annotate(txt, (idades_scaled[i], salarios_scaled[i]), textcoords="offset points", xytext=(0,10), ha='center')
plt.axvline(0, color='gray', linestyle='--', linewidth=0.8, label='Média Zero')
plt.axhline(0, color='gray', linestyle='--', linewidth=0.8)
plt.legend()
plt.show()
```
