```python
import numpy  as np
import pandas as pd
from   sklearn.model_selection import train_test_split
from   sklearn.preprocessing   import StandardScaler
from   sklearn.neighbors       import KNeighborsClassifier
from   sklearn.metrics         import accuracy_score, classification_report, confusion_matrix

dados = pd.read_excel('D:/PYTHON/01-CIENCIA DE DADOS/KNN - Nearest Neighbors/dados_gripe.xlsx')
dados.head(5)

```

```python
# Parâmetros do Algorítimo
K = 25
var_test_size = 0.30
var_random_state = 42


# novo_dado = np.array([[38.9,1,1,1,1,1,0,1]]) # GRIPADO   = ACERTO!!
# novo_dado = np.array([[38.9,1,1,1,1,1,0,1]]) # GRIPADO   = ACERTO!!
# novo_dado = np.array([[39.3,1,1,1,1,1,1,0]]) # GRIPADO   = ACERTO!!
novo_dado = np.array([[36.5,1,0,0,0,0,0,1]])  # SEM GRIPE = ACERTO!!


print(K)
```

```python
# --- Preparação dos Dados X ---
dados_features = [  'febre',
                    'tosse',
                    'dor_garganta',
                    'dor_corpo',
                    'dor_cabeça',
                    'mal_estar',
                    'calafrios',
                    'coriza' ]

X = dados[dados_features].values

# --- Preparação dos Dados Y ---
dados_rotulos =  ['gripe']
y = dados[dados_rotulos].values

print("X = ",X.shape, "y = ",y.shape)


# --- Divisão dos Dados ---
"""
test_size=0.25  = O total de 25% dos dados será usado para Teste, o restante para Treino.
test_size= 30   = O total de 30 registros dos dados será usado para Teste, o restante para Treino.

train_size=0.70 = O total de 30% dos dados será usado para Treinamento, o restante para Teste.
train_size= 60  = O total de 60 registros dos dados será usado para Treinamento, o restante para Teste.
"""
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
X_train[:10]

# --- Escalonamento dos Dados ---
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# --- Criação e Treinamento do Modelo KNN ---
knn_model = KNeighborsClassifier(n_neighbors=K)
knn_model.fit(X_train_scaled, y_train)

# --- Realização de Previsões ---
y_pred = knn_model.predict(X_test_scaled)

# --- Avaliação do Desempenho do Modelo ---
accuracy = accuracy_score(y_test, y_pred)
print(f"Acurácia do Modelo: {accuracy:.4f}\n")

print("Relatório de Classificação:\n", classification_report(y_test, y_pred))

print("Matriz de Confusão:\n", confusion_matrix(y_test, y_pred))


novo_dado_escalado = scaler.transform(novo_dado)
previsao_novo_dado = knn_model.predict(novo_dado_escalado)

print(f"\nPrevisão para o novo dado {novo_dado[0]}: Classe {previsao_novo_dado[0]}")

if previsao_novo_dado[0] == 1:
    print("O Paciente está com Gripe")
else:
    print("O Paciente não está com Gripe")
```
